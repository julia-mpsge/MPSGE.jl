# M63

Two country large group monopolistic competition

```julia
using MPSGE_MP
```

```julia
M63 = MPSGEModel()

@parameters(M63, begin
    ENDOWIL, 1
    ENDOWJL, 1
    ENDOWIS, 1
    ENDOWJS, 1
    TC, 1
    SUBSIDY, 0
    SIGMA, 5
end)

@sectors(M63, begin
 WFI   
 WFJ
 XI
 XII
 XIJ
 XJ
 XJI
 XJJ
 YI    
 YJ
 NI    
 NJ
end)

@commodities(M63, begin
 PY
 PUI   
 PUJ
 WI    
 WJ
 ZI    
 ZJ
 PXII   
 PXJI   
 PXJJ   
 PXIJ
 PXI
 PXJ
 FCI   
 FCJ
end)

@consumers(M63, begin
 CONSI   
 CONSJ
 ENTI    
 ENTJ
end)

@auxiliaries(M63, begin
 XQADJII
 XQADJJI
 XQADJJJ
 XQADJIJ
 XPADJI
 XPADJJ
end)




@production(M63, YI,[s=1, t=0], begin
	@output(PY, 100.0, t)
	@input(WI, 60.0, s)
	@input(ZI, 40.0, s)
end)

@production(M63, YJ,[s=1, t=0], begin
	@output(PY, 100.0, t)
	@input(WJ, 60.0, s)
	@input(ZJ, 40.0, s)
end)

@production(M63, XI,[s=1, t=0], begin
	@output(PXI, 80., t, taxes=[Tax(ENTI, .20)])
	@input(ZI, 48, s)
	@input(WI, 32, s)
end)

@production(M63, XII, [s=1, t=0], begin
	@output(PXII, 40., t, taxes=[Tax(CONSI, XPADJI*-1.0)])
	@input(PXI, 40., s)
end)

@production(M63, XIJ, [s=1, t=0], begin
	@output(PXIJ, (40/TC), t, taxes=[Tax(CONSJ, XPADJI*-1.0)])
	@input(PXI, (40.), s)
end)

@production(M63, XJ,[s=1, t=0], begin
	@output(PXJ, 80., t, taxes=[Tax(ENTJ, .20)])
	@input(ZJ, 48, s)
	@input(WJ, 32, s)
end)

@production(M63, XJI, [s=1, t=0], begin
	@output(PXJI, (40/TC), t, taxes=[Tax(CONSI, XPADJJ*-1.0)])
	@input(PXJ, (40.), s)
end)

@production(M63, XJJ, [s=1, t=0], begin
	@output(PXJJ, 40., t, taxes=[Tax(CONSJ, XPADJJ*-1.0)])
	@input(PXJ, 40., s)
end)

@production(M63, NI,[s=1, t=0], begin
	@output(FCI, 20, t)
	@input(ZI, 12, s)
	@input(WI, 8, s)
end)

@production(M63, NJ,[s=1, t=0], begin
	@output(FCJ, 20, t)
	@input(ZJ, 12, s)
	@input(WJ, 8, s)
end)

@production(M63, WFI,[s=1, t=0, a=>s=5], begin
	@output(PUI, 200., t)
	@input(PXII, 40., a, reference_price=1.25)
	@input(PXJI, 40, a, reference_price=1.25)
	@input(PY, 100., s)
end)

@production(M63, WFJ,[s=1, t=0, a=>s=5], begin
	@output(PUJ, 200., t)
	@input(PXJJ, 40., a, reference_price=1.25)
	@input(PXIJ, 40., a, reference_price=1.25)
	@input(PY, 100., s)
end)


@demand(M63, CONSI, begin
    @final_demand(PUI, 100*1+100*1+40*0 + 40*0)
end, begin
    @endowment(WI, 100*ENDOWIL)
    @endowment(ZI, 100*ENDOWIS)
    @endowment(PXII, 40*XQADJII)
    @endowment(PXJI, 40*XQADJJI)
end)

@demand(M63, CONSJ, begin
    @final_demand(PUJ, 100*1+100*1+40*0 + 40*0)
end, begin
    @endowment(WJ, 100*ENDOWJL)
    @endowment(ZJ, 100*ENDOWJS)
    @endowment(PXIJ, 40*XQADJIJ)
    @endowment(PXJJ, 40*XQADJJJ)
end)

@demand(M63, ENTI, begin
    @final_demand(FCI, 20)
end, begin

end)

@demand(M63, ENTJ, begin
    @final_demand(FCJ, 20)
end, begin

end)

@aux_constraint(M63, XQADJII,
	XQADJII - ( (NI^(.25))*XII - XII)
)

@aux_constraint(M63, XQADJJI,
	XQADJJI - ( (NJ^(.25))*XJI/TC - XJI/TC)
)

@aux_constraint(M63, XQADJJJ,
	XQADJJJ - ( (NJ^(.25))*XJJ - XJJ)
)

@aux_constraint(M63, XQADJIJ,
	XQADJIJ - ( (NI^(.25))*XIJ/TC - XIJ/TC)
)

@aux_constraint(M63, XPADJI,
	XPADJI - ( NI^.25 - 1)
)

@aux_constraint(M63, XPADJJ,
	XPADJJ - ( NJ^.25 - 1)
);
```

# Benchmark

```julia
using JuMP

JuMP.delete_lower_bound(get_variable(XQADJII))
JuMP.delete_lower_bound(get_variable(XQADJJI))
JuMP.delete_lower_bound(get_variable(XQADJIJ))
JuMP.delete_lower_bound(get_variable(XQADJJJ))

JuMP.delete_lower_bound(get_variable(XPADJI))
JuMP.delete_lower_bound(get_variable(XPADJJ))
```

```julia
fix(PY,1)

set_start_value(PXII, 1.25)
set_start_value(PXIJ, 1.25)
set_start_value(PXJI, 1.25)
set_start_value(PXJJ, 1.25)
set_start_value(PXI, 1.25)
set_start_value(PXJ, 1.25)


solve!(M63, cumulative_iteration_limit=0)
generate_report(M63)
```
